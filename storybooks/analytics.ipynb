```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
dataset = pd.read_csv("../input/dataset.csv")
dataset.head()
```

```python
sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})

tem = dataset.groupby(['erythema', 'disease']).size()
print(tem.shape)
tem = tem.reset_index()  
print(tem.columns)
tem.rename(columns = {0:'size'}, inplace = True)
sns.relplot("erythema", "disease", size="size", sizes=(15, 200), hue="size", palette="ch:r=-.5,l=.75", data=tem,  legend = False);
```
