## Erythemato Squamous

Erythemato squamous is a type of skin diseases in which diagonising has been hard for a long time. These types of skin diseases have **similar** or only a **very little** differences in test results. Hence, manual diagonising is failing us.

We are proposing a model with larger combinations and complexity to diagonise **six** types of erythemato-squamous diseases which are **psoriasis**, **seboreic dermatitis**, **lichen planus**, **pityriasis rosea**, **cronic dermatitis** and **pityriasis rubra pilaris**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

dataset = pd.read_csv("../input/dataset.csv")
dataset.head()
```

## How data is provided?
All data are in **ordinal data type** just because of the inaccuracy we get through the test results. We have to process the ordinal data which mostly likely falling under a limit of **0 to 3**. This limit indicates the **severity** or **intensity** of the test or a symptom of particular feature. The intensity is calculated as (0-Low --> 3-Extreme)

```python
sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})

tem = dataset.groupby(['erythema', 'disease']).size()
print(tem.shape)
tem = tem.reset_index()  
print(tem.columns)
tem.rename(columns = {0:'size'}, inplace = True)
sns.relplot("erythema", "disease", size="size", sizes=(15, 200), hue="size", palette="ch:r=-.5,l=.75", data=tem,  legend = False);
```
